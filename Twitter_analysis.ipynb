```python
def deltas (df):
    Delta_Adj_Close = []
    Delta_Volume = []
    df_delta = pd.DataFrame()
    
    for i in range(len(df['Dates'])-1):
        Delta_Adj_Close.append(df['Adj Close'][i+1]-df['Adj Close'][0])
        Delta_Volume.append(df['Volume'][i+1]-df['Volume'][0])
    
    df_delta = pd.DataFrame(
            {'Delta Adj Close':Delta_Adj_Close,
             'Delta Volume': Delta_Volume}
                ) 

    
    return (df_delta)
    
```
